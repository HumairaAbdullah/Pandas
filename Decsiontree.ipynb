```python
import pandas as pd 
```

```python
import numpy as np 
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv('C:\\Users\\DevValley\\Downloads\\play_prediction_weather.csv')
```

```python
data
```

```python
Outlook=LabelEncoder()
Temperature=LabelEncoder()
Humidity=LabelEncoder()	
Wind=LabelEncoder()
Play=LabelEncoder()
```

```python
data['Outlook']=Outlook.fit_transform(data['Outlook'])
data['Temperature']=Temperature.fit_transform(data['Temperature'])
data['Humidity']=Humidity.fit_transform(data['Humidity'])
data['Wind']=Wind.fit_transform(data['Wind'])
data['Play']=Play.fit_transform(data['Play'])
```

```python
data
```

```python
x=data[['Outlook','Temperature','Humidity','Wind']]
```

```python
x
```

```python
y=data['Play']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clasifier=DecisionTreeClassifier(criterion='entropy')
```

```python
clasifier.fit(x_train,y_train)
```

```python
clasifier.predict(x_test)
```

```python
x_test
```

```python
clasifier.score(x_test,y_test)
```

```python
from sklearn import tree
tree.plot_tree(clasifier)
```
